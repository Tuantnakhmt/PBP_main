### 4.4 LL_SMS_T

```python
Q1 = df['LL_SMS_T'].quantile(0.25)
Q3 = df['LL_SMS_T'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_filtered = df[(df['LL_SMS_T'] >= lower_bound) & (df['LL_SMS_T'] <= upper_bound)]

bin_width = 2 
num_bins = int((df_filtered['LL_SMS_T'].max() - df_filtered['LL_SMS_T'].min()) / bin_width)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

# bin_width = 50 
# num_bins = int((df['LL_SMS_T'].max() - df['LL_SMS_T'].min()) / bin_width)

sns.histplot(data=df_filtered, x='LL_SMS_T', hue='label', kde=False, element="step", stat="count", bins = num_bins ,common_norm=False)

plt.title("Phân phối LL SMS khả dụng tháng 9 của 2 nhóm (đã loại outlier)")
plt.xlabel("LL SMS khả dụng tháng 9")
plt.ylabel("Số lượng thuê bao")

plt.show()
```

## 6. Tương quan chung

### 6.1 Tương quan Số dư tháng 8 và độ giảm mức tiền nạp tháng 8-9

```python
def plot_correlation_with_regression(X, col1, col2):
    X = X[[col1, col2]].dropna()

    plt.figure(figsize=(10, 6))
    sns.regplot(x=col1, y=col2, data=X, scatter_kws={'alpha':0.6, 'color':'skyblue'}, line_kws={'color':'red'}, ci=None)
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title(f"Tương quan: {col1} và {col2}")
    plt.show()

plot_correlation_with_regression(df, 'So_du_T_1', 'decrease_NAP_THE_TvsT_1')
```

### 6.2 Tương quan giữa độ giảm số lần nạp t7-t6 và độ giảm mức tiền nạp t7-t8

```python
plot_correlation_with_regression(df, 'decrease_amount_nap_the_t7_t6', 'decrease_NAP_THE_TvsT_1')
```
